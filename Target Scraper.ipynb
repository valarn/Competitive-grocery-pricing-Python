```python
## This line of the scraper will go through each of the main categories in the grocery section and extracts the links
# for each one.


url = 'https://www.target.com/c/grocery/-/N-5xt1aZcl92v'
# url = "https://www.target.com/c/same-day-delivery/-/N-bswkz"
driver.get(url)
#Go to the list of Grocery items

# EC.presence_of_element_located((By.XPATH, '//*[@id="N-5xt1a"]/div/header/div/a'))
# driver.execute_script("window.scrollTo(0,400);")
# grocery_all = driver.find_element_by_xpath('//*[@id="N-5xt1a"]/div/header/div/a')
# grocery_all.click()




#Expand the list of the categories
sleep(2)
#EC.presence_of_element_located((By.XPATH, "//button[@data-test='showAllBtn']"))
driver.execute_script("window.scrollTo(0,400);")
# show_more = driver.find_elements_by_class_name('h-text-sm')
show_more = driver.find_element_by_xpath("//button[@data-test='showAllBtn']")
show_more.click()



# Recursively click on the categories
EC.presence_of_element_located((By.CSS_SELECTOR, ".ItemTitle-yw3r8r-0.eZUBZY"))
driver.execute_script("window.scrollTo(0,400);")
soup_cat = BeautifulSoup(driver.page_source,'lxml')
category_list = soup_cat.find_all('a',{'class':'Link-sc-1khjl8b-0 PicNavFilmstripItem-sc-1lmwnri-0 YutWE'})
category_links=[]
for category in category_list[:13]:
    category_links.append(category['href'])
    
```

```python
category_links
```

```python
# Each category of products comes with smaller sub categories that we need to go through before getting to the actual products themselves.

messy_subcategory_links=[]
for category_link in category_links:
    driver.get('https://www.target.com/c/'+category_link)
    EC.presence_of_element_located((By.CSS_SELECTOR, ".ItemTitle-yw3r8r-0.eZUBZY"))
    driver.execute_script("window.scrollTo(0,400);") 
    try:
        show_more = driver.find_element_by_xpath("//button[@data-test='showAllBtn']")
        show_more.click()
        
    except:
        pass
        
    soup_subcat = BeautifulSoup(driver.page_source,'lxml')
    subcategory_list = soup_subcat.find_all('a',{'class': 'Link-sc-1khjl8b-0'})
    for subcategory in subcategory_list:
       
        messy_subcategory_links.append(subcategory['href'])
        
# Removing the duplicate or extra links that were added to the messy_subcategory_links
dup_subcategory_links =[i for i in messy_subcategory_links if 'grocery/' in i.split('-') and i[:3]=="/c/"]
subcategory_links = list(dict.fromkeys(dup_subcategory_links))

subcategory_links_df = pd.DataFrame(subcategory_links)
subcategory_links_df.to_csv('datasets/subcategory_links.csv',mode = 'a' , header=False, index=False)
```

```python
## Links for all of the products
count_sub = 0
categories_to_scrape = dup_category_links2
for subcategory_link in categories_to_scrape: 
    driver.get('https://www.target.com' + subcategory_link+f'?Nao={0}')    
    count_sub += 1
    print(f'Scraping {count_sub} out of {len(categories_to_scrape)} subcategories')
    #Scrolling thru all pages
    driver.execute_script(f"window.scrollTo(0, document.body.scrollHeight);")
    driver.execute_script(f"window.scrollTo(0, document.body.scrollHeight-1000);")
    EC.presence_of_element_located((By.CLASS_NAME,"Col-favj32-0 SelectBox__ReactiveTextCol-sc-6gt3w9-0 eaQqfZ"))
    sleep(3)
    
    product_soup = BeautifulSoup(driver.page_source,'lxml')
    try:
        number_of_pages_html = product_soup.find('div',{'class':"Col-favj32-0 SelectBox__ReactiveTextCol-sc-6gt3w9-0 eaQqfZ"})
        number_of_pages = int(number_of_pages_html.text[-2:])
    except:
        number_of_pages
    
    
    for count in range(1,number_of_pages):
        product_list =[]
        print(f'Scraping {count} out of {number_of_pages} product pages!')
        
        #BeginSimulating mouse moves on the screen
        action = ActionChains(driver)
        hover_element= driver.find_elements_by_class_name("Button-bwu3xu-0")
        hover = action.move_to_element(hover_element[-2])
        hover.perform()
        action.move_by_offset(10,500)
        action.click()
        action.perform()
        action.move_by_offset(1000,400);
        action.click()
        action.perform()
        # End of simulation


        product_soup = BeautifulSoup(driver.page_source,'lxml')
        product_list = product_soup.find_all("a",{"data-test":"product-title"})
        #Adds all of the product links from one page to the list
        for product in product_list:
            product_links.append(product['href'])
            print(f"link scraped is: {product['href']}")
    
        try:
            if count == 1: 
                next_url = driver.current_url[:-1]+f'{(count)*24}'
                driver.get(next_url)
            else:
                next_url = driver.current_url[:-2]+f'{(count)*24}'
                driver.get(next_url)
        except:
               pass

   
    
    
        
```
